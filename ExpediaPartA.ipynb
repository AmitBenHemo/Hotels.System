## Set path for webdriver

```python
def move_to_element(driver, element):
    action = ActionChains(driver)
    action.move_to_element(element).perform()

path = 'C:\Windows\chromedriver_win32\chromedriver.exe'
#Open the driver
driver = webdriver.Chrome(path)
today=datetime.date.today()
updated_today = today.strftime("%Y-%m-%d")
tomorrow = today + datetime.timedelta(days=1)
updated_tomorrow = tomorrow.strftime("%Y-%m-%d")
website= f"https://www.expedia.com/Hotel-Search?adults=2&d1={updated_today}&d2={updated_tomorrow}&destination=New%20York%20%28and%20vicinity%29%2C%20New%20York%2C%20United%20States%20of%20America&endDate={updated_tomorrow}&latLong=40.75668%2C-73.98647&regionId=178293&rooms=1&semdtl=&sort=RECOMMENDED&startDate={updated_today}&theme=&useRewards=false&userIntent="
# TTT= time to travel
#los = length of stay

checkin = website.split("startDate=")[1].split("&")[0]
checkout = website.split("endDate=")[1].split("&")[0]
checkin_date = datetime.datetime.strptime(checkin, "%Y-%m-%d").date()
print(checkin_date)
updated_checkin = checkin_date.strftime("%Y-%m-%d")
print(updated_checkin)

df={
    "check-in":[],
    "check-out":[],
    "TTT": [],
    "LOS": [],
    "Index": [],
    "Hotel Name": [],
    "Price": [],
    "priceBeforesale": [],
    "Reviews": [],
    "Locatins": [],
    "Score": [],
    "Score_in_the_description": [],
    "Cancellation": [],
    "Includes": [],
    }
for TTT in range(0, 30):
    print(TTT)
    if(TTT>=1):
        checkin_date = checkin_date + datetime.timedelta(days=1)
        print(checkin_date)
        updated_checkin = checkin_date.strftime("%Y-%m-%d")
        print(updated_checkin)
    for LOS in range(1, 6):
        if(TTT==0):
            print(LOS)
            checkout_date = checkin_date + datetime.timedelta(days=LOS)
            print(checkout_date)
            updated_checkout = checkout_date.strftime("%Y-%m-%d")
            print(updated_checkout)
            checkin = website.split("startDate=")[1].split("&")[0]
            checkout = website.split("endDate=")[1].split("&")[0]
            updated_url = website.replace(checkin, updated_checkin).replace(checkout, updated_checkout)
            print(updated_url)
            driver.get(updated_url)
            sleep(2)
            print(driver.current_url)
            hotel_button= driver.find_elements(By.XPATH,'.//*[@class="uitk-card uitk-card-roundcorner-all uitk-card-has-primary-theme"]')
        else:
            print(LOS)
            checkout_date = checkin_date + datetime.timedelta(days=LOS)
            print(checkout_date)
            updated_checkout = checkout_date.strftime("%Y-%m-%d")
            print(updated_checkout)
            checkin = updated_url.split("startDate=")[1].split("&")[0]
            checkout = updated_url.split("endDate=")[1].split("&")[0]
            updated_url = updated_url.replace(checkin, updated_checkin).replace(checkout, updated_checkout)
            print(updated_url)
            driver.get(updated_url)
            sleep(2)
            print(driver.current_url)
            hotel_button= driver.find_elements(By.XPATH,'.//*[@class="uitk-card uitk-card-roundcorner-all uitk-card-has-primary-theme"]')  
        j=0
        driver.maximize_window()
        print(len(driver.find_elements(By.XPATH,'.//*[@class="uitk-card uitk-card-roundcorner-all uitk-card-has-primary-theme"]')))
        
        if(len(driver.find_elements(By.XPATH,'.//*[@class="uitk-card uitk-card-roundcorner-all uitk-card-has-primary-theme"]'))<100):
#             sleep(2)
#             move_to_element(driver, driver.find_element(By.XPATH,'//*[@id="app-layer-base"]/div[1]/main/div/div/div/div/div[2]/section[2]/div/div[2]/div/div[2]/div[1]/section/button'))
#             next_button= driver.find_element(By.XPATH,'//*[@id="app-layer-base"]/div[1]/main/div/div/div/div/div[2]/section[2]/div/div[2]/div/div[2]/div[1]/section/button')
#             sleep(2)
#             next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="app-layer-base"]/div[1]/main/div/div/div/div/div[2]/section[2]/div/div[2]/div/div[2]/div[1]/section/button')))
#             next_button.click()
#             sleep(7)
#             driver.execute_script("window.scrollTo(3,document.body.scrollHeight)")
            driver.execute_script("window.scrollTo(2,document.body.scrollHeight)")
            driver.execute_script("window.scrollTo(2,document.body.scrollHeight)")
            driver.execute_script("window.scrollTo(2,document.body.scrollHeight)")
            
        else:
            driver.execute_script("window.scrollTo(3,document.body.scrollHeight)")
            sleep(3)



        for curHotel in driver.find_elements(By.XPATH,'.//*[@class="uitk-card uitk-card-roundcorner-all uitk-card-has-primary-theme"]'):
            j=j+1
            df["Index"].append(j)
            df["TTT"].append(TTT+1)
            df["LOS"].append(LOS)
            df["check-in"].append(updated_checkin)
            df["check-out"].append(updated_checkout)
            try:
                df["Hotel Name"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-heading uitk-heading-5 overflow-wrap"]').text)
            except:
                df["Hotel Name"].append(None)


            try:
                df["Score"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-spacing uitk-spacing-padding-inlineend-half uitk-layout-flex-item"]').text)
            except:
                df["Score"].append(None)


            try:
                df["Score_in_the_description"].append(curHotel.find_elements(By.XPATH,'.//*[@class="uitk-text uitk-type-300 uitk-text-default-theme"]')[1].text)
            except:
                df["Score_in_the_description"].append(None)
            
            print('the len of reviews- ', len(curHotel.find_elements(By.XPATH,'.//*[@class="uitk-text uitk-type-300 uitk-text-default-theme"]')))
            try:
                df["Reviews"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-layout-flex uitk-layout-flex-align-items-flex-start"]').text.split("\n")[3])
            except:
                df["Reviews"].append("0")
            
            try:
                df["priceBeforesale"].append(curHotel.find_elements(By.XPATH,'.//*[@class="uitk-text uitk-type-300 uitk-text-default-theme"]')[2].text)
            except:
                df["priceBeforesale"].append("0")

            try:
                df["Price"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-text uitk-type-end uitk-type-200 uitk-text-default-theme"]').text)
            except:
                df["Price"].append(None)


            try:
                df["Locatins"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-text uitk-text-spacing-half truncate-lines-2 uitk-type-300 uitk-text-default-theme"]').text)
            except:
                df["Locatins"].append(None)


            try:
                df["Includes"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-text truncate uitk-type-200 uitk-text-default-theme"]').text)
            except:
                df["Includes"].append(None)


            try:
                df["Cancellation"].append(curHotel.find_element(By.XPATH,'.//*[@class="uitk-text uitk-type-300 uitk-text-positive-theme"]').text)
            except:
                df["Cancellation"].append('0')

        driver.quit()
        driver = webdriver.Chrome(path)
        print("Done Scraping",TTT, LOS)
```
