## Import Modules

```python
import http
import pandas as pd
from IPython.display import display, HTML
import time
from time import sleep
import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

## Set path for webdriver

```python
path = 'C:\Windows\chromedriver_win32\chromedriver.exe'
#Open the driver
driver = webdriver.Chrome(path)
website= 'https://www.booking.com/searchresults.html?ss=New+York%2C+New+York+State%2C+United+States&ssne=%D7%A0%D7%99%D7%95+%D7%99%D7%95%D7%A8%D7%A7&ssne_untouched=%D7%A0%D7%99%D7%95+%D7%99%D7%95%D7%A8%D7%A7&sid=6ee229e721427ded7a64cee1f3f14704&aid=7946844&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=20088325&dest_type=city&ac_position=0&ac_click_type=b&ac_langcode=en&ac_suggestion_list_length=5&search_selected=true&search_pageview_id=fa7c3e7d446f0313&ac_meta=GhBmYTdjM2U3ZDQ0NmYwMzEzIAAoATICZW46A25ld0AASgBQAA%3D%3D&checkin=2023-03-08&checkout=2023-03-09&group_adults=2&no_rooms=1&group_children=0&sb_travel_purpose=leisure&soz=1&lang_changed=1'
# TTT= time to travel
#los = length of stay

checkin = website.split("checkin=")[1].split("&")[0]
checkout = website.split("checkout=")[1].split("&")[0]
checkin_date = datetime.datetime.strptime(checkin, "%Y-%m-%d").date()
print(checkin_date)
updated_checkin = checkin_date.strftime("%Y-%m-%d")
print(updated_checkin)

df={
    "TTT": [],
    "LOS": [],
    "Hotel Name": [],
    "Price": [],
    "Reviews": [],
    "NumberOfNights": [],
    "Locatins": [],
    "Score": [],
    "Score_in_the_description": [],
    "priceBeforesale": [],
    "Cancellation": [],
    "prepayment": [],
    "typeBed": [],
    "DistanceFromCenter": [],
    "Subway Access": [],
    "Stars": []
    }
for TTT in range(0, 30):
    print(TTT)
    if(TTT>=1):
        checkin_date = checkin_date + datetime.timedelta(days=1)
        print(checkin_date)
        updated_checkin = checkin_date.strftime("%Y-%m-%d")
        print(updated_checkin)
    for LOS in range(1, 6):
        if(TTT==0):
            print(LOS)
            checkout_date = checkin_date + datetime.timedelta(days=LOS)
            print(checkout_date)
            updated_checkout = checkout_date.strftime("%Y-%m-%d")
            print(updated_checkout)
            checkin = website.split("checkin=")[1].split("&")[0]
            checkout = website.split("checkout=")[1].split("&")[0]
            updated_url = website.replace(checkin, updated_checkin).replace(checkout, updated_checkout)
            print(updated_url)
            driver.get(updated_url)
            print(driver.current_url)
            hotel_button= driver.find_elements(By.XPATH,'.//*[@class="d20f4628d0"]')
        else:
            print(LOS)
            checkout_date = checkin_date + datetime.timedelta(days=LOS)
            print(checkout_date)
            updated_checkout = checkout_date.strftime("%Y-%m-%d")
            print(updated_checkout)
            checkin = updated_url.split("checkin=")[1].split("&")[0]
            checkout = updated_url.split("checkout=")[1].split("&")[0]
            updated_url = updated_url.replace(checkin, updated_checkin).replace(checkout, updated_checkout)
            print(updated_url)
            driver.get(updated_url)
            print(driver.current_url)
            hotel_button= driver.find_elements(By.XPATH,'.//*[@class="d20f4628d0"]') 

        for i in range(4):
            print("scraping page", i+1)
            for curHotel in driver.find_elements(By.XPATH,'.//*[@class="d20f4628d0"]'):
                df["TTT"].append(TTT+1)
                df["LOS"].append(LOS)
                try:
                    df["Hotel Name"].append(curHotel.find_element(By.XPATH,'.//*[@class="fcab3ed991 a23c043802"]').text)
                except:
                    df["Hotel Name"].append("Error")

                try:
                    df["Price"].append(curHotel.find_element(By.XPATH,'.//*[@class="fcab3ed991 fbd1d3018c e729ed5ab6"]').text.split(" ")[1])       
                except:
                    df["Price"].append("Error")  

                try:
                    df["Reviews"].append(curHotel.find_element(By.XPATH,'.//*[@class="d8eab2cf7f c90c0a70d3 db63693c62"]').text.split(" ")[0])       
                except:
                    df["Reviews"].append("0") 

                try:
                    df["NumberOfNights"].append(curHotel.find_element(By.XPATH,'.//*[@class="d8eab2cf7f c90c0a70d3"]').text.split()[0])       
                except:
                    df["NumberOfNights"].append(0) 

                try:
                    df["Locatins"].append(curHotel.find_element(By.XPATH,'.//*[@class="f4bd0794db b4273d69aa"]').text.split(",")[0])       
                except:
                    df["Locatins"].append("Error") 

                try:
                    df["Score"].append(curHotel.find_element(By.XPATH,'.//*[@class="b5cd09854e d10a6220b4"]').text)
                except:
                    df["Score"].append("0")

                try:
                    df["Score_in_the_description"].append(curHotel.find_element(By.XPATH,'.//*[@class="b5cd09854e f0d4d6a2f5 e46e88563a"]').text)
                except:
                    df["Score_in_the_description"].append("NaN") 

                try:
                    df["priceBeforesale"].append(curHotel.find_element(By.XPATH,'.//*[@class="c5888af24f e729ed5ab6"]').text.split(" ")[1])
                except:
                    df["priceBeforesale"].append("NaN") 

                try:
                    df["Cancellation"].append(curHotel.find_element(By.XPATH,'.//*[@class="d506630cf3"]').text.split()[0]) 
                except:
                    df["Cancellation"].append("Not Free") 

                try:
                    df["prepayment"].append(curHotel.find_element(By.XPATH,'.//*[@class="d506630cf3"]').text.split()[3]) 
                except:
                    df["prepayment"].append("Yes") 

                try:
                    df["typeBed"].append(curHotel.find_element(By.XPATH,'.//*[@class="df597226dd"]').text) 
                except:
                    df["typeBed"].append("Error") 

                try:
                    df["DistanceFromCenter"].append(curHotel.find_element(By.XPATH,'.//*[@class="a1fbd102d9"]').text.split("Show on map")[1].split("from")[0])       
                except:
                    df["DistanceFromCenter"].append("Error")

                try:
                    df["Subway Access"].append(curHotel.find_element(By.XPATH,'.//*[@class="a1fbd102d9"]').text.split("center")[1])       
                except:
                    df["Subway Access"].append("NaN")

                try:
                    df["Stars"].append(curHotel.find_element(By.XPATH,'.//*[@class="e4755bbd60"]').get_attribute("aria-label")[0])  
                except:
                    df["Stars"].append("Error")
                
            try:
                next_button= driver.find_element(By.XPATH,".//*[@class='f32a99c8d1 f78c3700d2']")
                next_button.click()
                sleep(2)
            except:
                continue
        print("Done Scraping",TTT, LOS)
```
